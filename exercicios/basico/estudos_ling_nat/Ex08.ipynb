```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Gerando os dados de chuva e colheita (valores j fornecidos)
desvio = np.random.random(size=100)
mm_chuva = np.arange(50, 150) * (desvio * 0.10)
colheita = np.arange(3, 103) * (desvio * 0.10)

# Inicializando variveis para armazenar os melhores parmetros e o menor custo
melhor_B = None
melhor_W = None
menor_custo = float('inf')

# Realizando 10 tentativas
for _ in range(10):
    # Gerando valores aleatrios para B e W
    B = np.random.uniform(-10, 10)
    W = np.random.uniform(-10, 10)

    # Calculando a previso da colheita com base na quantidade de chuva usando a funo linear
    previsao_colheita = B + W * mm_chuva

    # Calculando o custo (erro quadrtico mdio)
    custo = np.mean((previsao_colheita - colheita) ** 2)

    # Verificando se este custo  menor que o menor custo at agora
    if custo < menor_custo:
        melhor_B = B
        melhor_W = W
        menor_custo = custo

        # Salvando os valores para a melhor tentativa para plotagem posterior
        melhor_previsao_colheita = previsao_colheita

# Plotando os pontos de dados de chuva e colheita
plt.scatter(mm_chuva, colheita, label='Dados Observados')

# Plotando as linhas de regresso linear para a melhor tentativa
plt.plot(mm_chuva, melhor_previsao_colheita, color='red', label='Regresso Linear (Melhor Tentativa)')

plt.xlabel('Quantidade de Chuva (mm)')
plt.ylabel('Quantidade de Colheita')
plt.title('Relao entre Chuva e Colheita')
plt.legend()
plt.grid(True)
plt.show()
```
